#3.Feature Understanding
(Univariate Analysis)
plotting feature distributions - Histogram | KDE | Boxplot


```python
#count by year
df['Year_Introduced'].value_counts()
```

```python
#number of roller coasters introduced in first 10 years
ax = df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind = 'bar', title = 'Number of Roller Coasters Introduced by Year')

ax.set_xlabel('Year Introduced')
ax.set_ylabel('Number of Roller Coaster')
```

```python
#speed Hist
ax = df['Speed_Mph'].plot(kind='hist',
                          bins =20,
                          title='Coaster Speed (mph)')

ax.set_xlabel('Speed (mph)')
```

```python
#speed KDE
ax = df['Speed_Mph'].plot(kind='kde',
                          title='Coaster Speed (mph)')

ax.set_xlabel('Speed (mph)')
```

```python
#Gforce
ax = df['Gforce'].plot(kind='hist',
                          bins =35,
                          title='Gforce')

ax.set_xlabel('Gforce')
```

```python
#Gforce kde
ax = df['Gforce'].plot(kind='kde',
                          title='Gforce')

ax.set_xlabel('Gforce')
```

```python
df['Type_Main'].value_counts()
```

#Future Relationships
Scatterplots, Heatmap Correlations, pairplots, Groupby Comparisons

```python
#scatter speed vs height
df.plot(kind='scatter',
        x='Speed_Mph',
        y='Height_ft',
        title='Coaster Speed vs. Height')
plt.show()
```

```python
#seaborn scatter
ax = sns.scatterplot(x='Speed_Mph',
                y='Height_ft',
                hue='Year_Introduced',
                data=df)
ax.set_title('Coaster Speed vs. Height')
plt.show()
```

```python
#seaborn pairplot
sns.pairplot(df,
             vars=['Year_Introduced','Speed_Mph',
                   'Height_ft','Inversions','Gforce'],
            hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced','Speed_Mph',
    'Height_ft','Inversions','Gforce']].dropna().corr()
df_corr
```

```python
#Correlation in a sns heatmap
sns.heatmap(df_corr, annot=True)
```

What are the locations with the fastest roller coasters(minimum 10 at the location)?

```python
ax = df.query('Location != "Other"') \
  .groupby('Location')['Speed_Mph'] \
  .agg(['mean','count']) \
  .query('count >= 10') \
  .sort_values('mean')['mean']\
  .plot(kind = 'barh', figsize = (12,5), title = 'Average Coaster Speed by Location')

ax.set_xlabel('Average Coaster Speed')
ax.set_ylabel('Location')
plt.show()
```
